## Average Total Poverty Rate between City Types

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data12 = pd.read_csv('Race_Data/2012_Race_Data.csv')
data13 = pd.read_csv('Race_Data/2013_Race_Data.csv')
data14 = pd.read_csv('Race_Data/2014_Race_Data.csv')
data15 = pd.read_csv('Race_Data/2015_Race_Data.csv')
data16 = pd.read_csv('Race_Data/2016_Race_Data.csv')
data17 = pd.read_csv('Race_Data/2017_Race_Data.csv')
```

```python
data12 = data12.mean()
data13 = data13.mean()
data14 = data14.mean()
data15 = data15.mean()
data16 = data16.mean()
data17 = data17.mean()
```

```python
poverty = ["Total Poverty Rate (%)", "White Poverty Rate (%)", "Black Poverty Rate (%)", 
           "Asian Poverty Rate (%)", "Hispanic Poverty Rate (%)"]


race_df = pd.DataFrame({"Poverty Type": poverty,
                        "2012": data12, "2013": data13,
                        "2014": data14, "2015": data15, 
                        "2016": data16, "2017": data17})
```

```python
years = [2012,2013,2014,2015,2016,2017]

total, = plt.plot(years, race_df.loc['Total Poverty Rate (%)',["2012","2013","2014","2015","2016","2017"]], 
                        color="blue",label=race_df.loc['Total Poverty Rate (%)',"Poverty Type"])

white, = plt.plot(years, race_df.loc['White Poverty Rate (%)',["2012","2013","2014","2015","2016","2017"]], 
                        color="green",label=race_df.loc['White Poverty Rate (%)',"Poverty Type"])

black, = plt.plot(years, race_df.loc['Black Poverty Rate (%)',["2012","2013","2014","2015","2016","2017"]], 
                        color="red",label=race_df.loc['Black Poverty Rate (%)',"Poverty Type"])

asian, = plt.plot(years, race_df.loc['Asian Poverty Rate (%)',["2012","2013","2014","2015","2016","2017"]], 
                        color="orange",label=race_df.loc['Asian Poverty Rate (%)',"Poverty Type"])

hispanic, = plt.plot(years, race_df.loc['Hispanic Poverty Rate (%)',["2012","2013","2014","2015","2016","2017"]], 
                        color="brown",label=race_df.loc['Hispanic Poverty Rate (%)',"Poverty Type"])

plt.legend(handles=[total, white, black, asian, hispanic], loc="best", prop={'size': 6})
plt.xlabel("Years")
plt.ylabel("Poverty Rate (%)")
plt.title("Poverty Rate Between 2012-2017 By Race")
plt.show()
```
