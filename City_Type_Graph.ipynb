## Average Total Poverty Rate between City Types

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data12 = pd.read_csv('Poverty_Data/2012_Poverty_Data.csv')
data13 = pd.read_csv('Poverty_Data/2013_Poverty_Data.csv')
data14 = pd.read_csv('Poverty_Data/2014_Poverty_Data.csv')
data15 = pd.read_csv('Poverty_Data/2015_Poverty_Data.csv')
data16 = pd.read_csv('Poverty_Data/2016_Poverty_Data.csv')
data17 = pd.read_csv('Poverty_Data/2017_Poverty_Data.csv')
```

```python
bins = [0, 24000, 99999, 500000]
city_type = ["Rural", "Suburban", "Urban"]

data12["City Type"] = pd.cut(data12["Population"],bins,labels=city_type)
city_type12 = data12.groupby("City Type")
avg_poverty12 = city_type12["Total Poverty Rate (%)"].mean()

data13["City Type"] = pd.cut(data13["Population"],bins,labels=city_type)
city_type13 = data13.groupby("City Type")
avg_poverty13 = city_type13["Total Poverty Rate (%)"].mean()

data14["City Type"] = pd.cut(data14["Population"],bins,labels=city_type)
city_type14 = data14.groupby("City Type")
avg_poverty14 = city_type14["Total Poverty Rate (%)"].mean()

data15["City Type"] = pd.cut(data15["Population"],bins,labels=city_type)
city_type15 = data15.groupby("City Type")
avg_poverty15 = city_type15["Total Poverty Rate (%)"].mean()

data16["City Type"] = pd.cut(data16["Population"],bins,labels=city_type)
city_type16 = data16.groupby("City Type")
avg_poverty16 = city_type16["Total Poverty Rate (%)"].mean()

data17["City Type"] = pd.cut(data17["Population"],bins,labels=city_type)
city_type17 = data17.groupby("City Type")
avg_poverty17 = city_type17["Total Poverty Rate (%)"].mean()

city_df = pd.DataFrame({"City Type": city_type,
                        "2012": avg_poverty12, "2013": avg_poverty13,
                        "2014": avg_poverty14, "2015": avg_poverty15, 
                        "2016": avg_poverty16, "2017": avg_poverty17})
```

```python
combined_rate = city_df.mean()
years = combined_rate.keys()

total_avg, = plt.plot(years, combined_rate, color="orange", label="Combined Average" )

rural, = plt.plot(years, city_df.loc['Rural',["2012","2013","2014","2015","2016", "2017"]], 
                        color="green",label=city_df.loc['Rural',"City Type"])

suburban, = plt.plot(years, city_df.loc['Suburban',["2012","2013","2014","2015","2016", "2017"]], 
                        color="red",label=city_df.loc['Suburban',"City Type"])

urban, = plt.plot(years, city_df.loc['Urban',["2012","2013","2014","2015","2016", "2017"]], 
                        color="blue",label=city_df.loc['Urban',"City Type"])

plt.legend(handles=[total_avg, rural, suburban, urban], loc="best", prop={'size': 6})
plt.xlabel("Years")
plt.ylabel("Poverty Rate (%)")
plt.title("Poverty Rate Between 2012-2017 By Population Type")
plt.show()
```
